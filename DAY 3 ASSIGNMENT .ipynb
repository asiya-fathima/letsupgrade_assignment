# 2 Use For Loop to Print Prime Numbers in between 1 to 1000

```python
for i in range(1, 1000):
    for j in range(2, i//2):
        if i % j == 0:
            break
    else:
        print("Prime Number", i)
```

# 3.Write a program for printing the tables from 1,10 using Nested For Loop

```python
for i in range(1,11):
    print("")
    for j in range(1,11):
        print(i,"x",j,"=",i*j)
```
